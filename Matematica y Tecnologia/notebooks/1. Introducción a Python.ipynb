## **3. Operadores: Las Herramientas del Cálculo**

Los operadores son símbolos especiales que realizan operaciones sobre los valores (operandos).

### **3.1. Operadores Aritméticos**

Son los que ya conocemos de la aritmética básica.

| Operador | Descripción              | Ejemplo Matemático       | Ejemplo en Python |
| :------: | :----------------------- | :----------------------- | :---------------- |
| `+`      | Suma                     | $a + b$                  | `a + b`           |
| `-`      | Resta                    | $a - b$                  | `a - b`           |
| `*`      | Multiplicación           | $a \times b$             | `a * b`           |
| `/`      | División                 | $\frac{a}{b}$            | `a / b`           |
| `//`     | División Entera (cociente) | $\lfloor a/b \rfloor$   | `a // b`          |
| `%`      | Módulo (resto)           | $a \pmod{b}$             | `a % b`           |
| `**`     | Exponenciación           | $a^b$                    | `a ** b`          |

```python
# Celda de Código 4: Operadores Aritméticos
a = 17
b = 5

print(f"{a} + {b} = {a + b}")
print(f"{a} / {b} = {a / b}")
print(f"El cociente de 17 dividido 5 es: {a // b}") # División entera
print(f"El resto de 17 dividido 5 es: {a % b}") # Módulo
print(f"{a} elevado a {b} es: {a ** b}")
```

### **3.2. Operadores Relacionales y de Asignación**

*   **Relacionales:** Comparan dos valores y devuelven un booleano (`True` o `False`). Son la base de la toma de decisiones. Incluyen `==` (igual a), `!=` (distinto de), `>` (mayor que), `<` (menor que), `>=` (mayor o igual que), `<=` (menor o igual que).

*   **Asignación:** Asignan un valor a una variable. El básico es `=`, pero existen combinaciones como `+=`, `-=`, `*=`, que realizan una operación y asignan el resultado en un solo paso. Por ejemplo, `x += 2` es un atajo para `x = x + 2`.

```python
felix=10**6
print(felix)
felix=felix+5
print(felix)
```

```python
# Celda de Código 5: Operadores Relacionales y de Asignación
x = 10
y = 12

# Relacionales
print(f"¿Es {x} igual a {y}? {x == y}")
print(f"¿Es {x} distinto de {y}? {x != y}")
print(f"¿Es {x} menor o igual que {y}? {x <= y}")
```

### **3.3. Operadores Lógicos**

Estos operadores (`and`, `or`, `not`) nos permiten combinar expresiones booleanas, fundamentales para la lógica proposicional.

*   `and`: Devuelve `True` si **ambas** expresiones son verdaderas.
*   `or`: Devuelve `True` si **al menos una** de las expresiones es verdadera.
*   `not`: Invierte el valor de verdad de una expresión.

```python
# Celda de Código 6: Operadores Lógicos
# Ejemplo: Verificar si un número 'n' está en el intervalo [0, 100]
n = 75
condicion1 = n >= 0
condicion2 = n <= 100

print(f"¿El número {n} está entre 0 y 100? {condicion1 and condicion2}")
```

### **4.2. Métodos Principales de Strings**

Los "métodos" son funciones que pertenecen a un objeto. En este caso, son acciones que podemos realizar sobre una cadena de texto.

*   `len()`: (Técnicamente una función, no un método) Devuelve la longitud de la cadena.
*   `.upper()` / `.lower()`: Convierten a mayúsculas / minúsculas.
*   `.strip()`: Elimina espacios en blanco al inicio y al final.

```python
# Celda de Código 8: Métodos de strings
frase = "  La Geometría es el arte de razonar bien sobre figuras mal dibujadas.  "
print("Frase original:", frase)
print("Longitud:", len(frase))
print("Sin espacios extra:", frase.strip())
frase2=frase.strip()
print("Longitud:", len(frase2))
print("En mayúsculas:", frase.upper())
print("Reemplazando 'Geometría' por 'Álgebra':", frase.replace("Geometría", "Álgebra"))

# Dividiendo la frase en palabras
palabras = frase.strip().split(" ")
print("Lista de palabras:", palabras)
```

## **7. Funciones: Modularizando Nuestro Código**

Una función es un bloque de código reutilizable que realiza una tarea específica. Nos ayudan a organizar el código, hacerlo más legible y evitar la repetición.

### **7.1. Funciones `def`**

Es la forma estándar de definir una función.

**Ejemplo Matemático 1:** Función para calcular el área de un círculo, $A = \pi r^2$.

```python
def miprimerafuncion(x,y,z):
  return x**2+5*y-z**2
```

```python
miprimerafuncion(2,4,3)
```

```python
def funcion_discriminante(a,b,c):
  d=b**2-4*a*c
  if d==0:
    print("Las raices son iguales y reales")
  elif d>0:
    print("Las raices son diferentes y reales")
  else:
    print("Las raices son complejas")

  return f"El discriminante es: {d}"
```

```python
funcion_discriminante(1,4,5)
```

```python
def funcionespecial(*x):
  suma=0
  for i in x:
    print(i)
    suma=suma+i
    print(suma)
  return suma
```

```python
funcionespecial(2,3,4,6,78,8)
```

```python
# Celda de Código 13: Función def para el área de un círculo
import math # Necesitamos la constante pi

def calcular_area_circulo(radio):
    """Calcula el área de un círculo dado su radio."""
    if radio < 0:
        return "El radio no puede ser negativo."
    area = math.pi * (radio ** 2)
    return area
```

```python
calcular_area_circulo(1)
```

**Ejemplo Matemático 2:** Función para calcular el factorial de un número, $n! = n \times (n-1) \times \dots \times 1$.

```python
# Celda de Código 14: Función def para el factorial
def factorial(n):
    """Calcula el factorial de un número entero no negativo."""
    if not isinstance(n, int) or n < 0:
        return "El factorial solo está definido para enteros no negativos."
    if n == 0:
        return 1

    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

num = 5
print(f"El factorial de {num} es {factorial(num)}")
```

### **7.2. Funciones `lambda`**

Son funciones anónimas y de una sola línea. Son útiles para tareas cortas y concisas.

**Ejemplo Matemático 1:** Una función para evaluar el polinomio $f(x) = x^2 + 2x - 1$.

```python
# Celda de Código 15: Función lambda para un polinomio
polinomio = lambda x: x**2 + 2*x - 1

print(f"El valor de f(x) en x=3 es: {polinomio(3)}")
```

```python
# Celda de Código 17: Importando y usando la librería math
import math

# 1. Constante Pi
print(f"1. Valor de pi: {math.pi}")

# 2. Constante e (número de Euler)
print(f"2. Valor de e: {math.e}")

# 3. Raíz cuadrada
print(f"3. Raíz cuadrada de 81: {math.sqrt(81)}")

# 4. Funciones trigonométricas (¡usan radianes!)
angulo_grados = 90
angulo_radianes = math.radians(angulo_grados)
print(f"4. Seno de {angulo_grados}° es: {math.sin(angulo_radianes)}")

# 5. Logaritmos
print(f"5. Logaritmo natural de 10: {math.log(10)}")
print(f"   Logaritmo en base 10 de 100: {math.log10(100)}")

# 6. Factorial (ya lo implementamos, pero math lo tiene)
print(f"6. Factorial de 6 (con math): {math.factorial(6)}")

# 7. Redondeo hacia arriba (techo) y hacia abajo (suelo)
num = 4.7
print(f"7. Techo de {num} es {math.ceil(num)}, Suelo de {num} es {math.floor(num)}")

# 8. Máximo Común Divisor (MCD)
print(f"8. MCD de 48 y 60 es: {math.gcd(48, 60)}")

# 9. Potencia (alternativa a **)
print(f"9. 2 elevado a la 10: {math.pow(2, 10)}")

# 10. Grados a partir de radianes
print(f"10. {math.pi/2} radianes son {math.degrees(math.pi/2)} grados.")
```

## **9. Un Paso Más Allá: List Comprehensions**

Las "List Comprehensions" o "comprensiones de listas" son una forma elegante y "Pythónica" de crear listas. Su sintaxis es muy similar a la notación de construcción de conjuntos en matemáticas.

**Ejemplo Matemático:** Crear una lista de los primeros 10 cuadrados perfectos.  
Notación matemática: $S = \{x^2 \mid x \in \{0, 1, 2, \dots, 9\}\}$

```python
# Celda de Código 18: List Comprehensions
# Forma tradicional con un bucle for
cuadrados_for = []
for x in range(10):
    cuadrados_for.append(x**2)
print("Lista con bucle for: ", cuadrados_for)


# Forma elegante con List Comprehension
cuadrados_lc = [x**2 for x in range(10)]
print("Lista con list comprehension:", cuadrados_lc)

# Ejemplo más complejo: una lista de números pares del 1 al 20
pares = [num for num in range(1, 21) if num % 2 == 0]
print("Números pares del 1 al 20:", pares)
```
