```python
from sympy import Function, Eq, dsolve, Derivative, symbols

# Definir las variables y la función desconocida
t = symbols('t')
x = Function('x')(t)

# Escribir la ecuación diferencial
ode = Eq(Derivative(x, t, t) + x, 0)

# Definir las condiciones iniciales
ics = {x.subs(t, 0): 1, x.diff(t).subs(t, 0): 0}

# Resolver la ecuación diferencial con las condiciones iniciales
sol = dsolve(ode, x, ics=ics)

# Mostrar la solución
sol
```

```python
from sympy import symbols, Function, Eq, Derivative, sin, pi, exp, solve, dsolve

# Definición de las variables y funciones
x, t, alpha = symbols('x t alpha')
u = Function('u')(x, t)

# Definición de la ecuación diferencial parcial
heat_eq = Eq(Derivative(u, t), alpha * Derivative(u, x, x))

# Mostrar la ecuación diferencial
print("Ecuación Diferencial Parcial:")
print(heat_eq)

# Resolver la ecuación con condiciones iniciales y de frontera
# Supongamos la forma de solución separando variables u(x, t) = X(x) * T(t)
X = Function('X')(x)
T = Function('T')(t)

# Ecuaciones para X(x) y T(t)
eq_X = Eq(Derivative(X, x, x), -lambda_ * X)
eq_T = Eq(Derivative(T, t), -lambda_ * alpha * T)

# Resolver la ecuación para T(t)
lambda_ = symbols('lambda')
sol_T = dsolve(eq_T, T)
print("\nSolución para T(t):")
print(sol_T)

# Resolver la ecuación para X(x)
sol_X = dsolve(eq_X, X)
print("\nSolución para X(x):")
print(sol_X)

# Construir la solución general
u_sol = sol_X.rhs * sol_T.rhs
print("\nSolución General de la EDP:")
print(u_sol)
```
